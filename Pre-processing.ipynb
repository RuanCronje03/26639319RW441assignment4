```python
import pandas as pd
spambase_df = pd.read_csv("spambase.csv")
mushroom_df = pd.read_csv("mushroom.csv")
segment_df  = pd.read_csv("segment.csv")
```

```python
for name, df in {
    "Spambase": spambase_df,
    "Mushroom": mushroom_df,
    "Segment": segment_df
}.items():
    target_col = "class" 
    n_classes = df[target_col].nunique()
    print(f"{name:10s} | Shape: {df.shape} | Classes: {n_classes}")
```

```python
for name, df in {
    "Spambase": spambase_df,
    "Mushroom": mushroom_df,
    "Segment": segment_df
}.items():
    missing_total = df.isna().sum().sum()
    cols_with_missing = df.columns[df.isna().any()].tolist()
    print(f"{name:10s} | Total missing values: {missing_total}")
    if cols_with_missing:
        print(f"Columns with missing data: {cols_with_missing}")
    else:
        print("No missing values")
```

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

datasets = {
    "spambase": spambase_df,
    "mushroom": mushroom_df,
    "segment":  segment_df
}

encoded_datasets = {}

for name, df in datasets.items():
    df_encoded = df.copy()

    # Fill missing values
    for col in df_encoded.columns:
        if df_encoded[col].dtype == "O" or str(df_encoded[col].dtype) == "category":
            df_encoded[col] = df_encoded[col].fillna("unknown")
        else:
            df_encoded[col] = df_encoded[col].fillna(df_encoded[col].mean())

    # encode all categorical columns 
    for col in df_encoded.columns:
        if df_encoded[col].dtype == "O" or str(df_encoded[col].dtype) == "category":
            le = LabelEncoder()
            df_encoded[col] = le.fit_transform(df_encoded[col])

    encoded_datasets[name] = df_encoded
    print(f"{name.capitalize():10s} | shape: {df_encoded.shape} | all numeric")

encoded_datasets["spambase"].to_csv("spambase_encoded.csv", index=False)
encoded_datasets["mushroom"].to_csv("mushroom_encoded.csv", index=False)
encoded_datasets["segment"].to_csv("segment_encoded.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
for i, (name, df) in enumerate(datasets.items(), 1):
    target_col = "class"
    counts = df[target_col].value_counts().sort_index()

    plt.subplot(1, 3, i)
    counts.plot(kind="bar", color="steelblue", edgecolor="black")
    plt.title(f"{name} — {target_col}")
    plt.xlabel("Class label")
    plt.ylabel("Frequency")
    plt.tight_layout()

plt.show()
```
