```python
results1 = pd.read_csv("df_results.csv")
results2 = pd.read_csv("df_results2.csv")
results3 = pd.read_csv("df_results3.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Replace NaN 
results1["max_depth"] = results1["max_depth"].fillna("None").astype(str)

numeric_depths = sorted(
    [float(d) for d in results1["max_depth"].unique() if d != "None"]
)
depth_order = [str(int(d)) if d.is_integer() else str(d) for d in numeric_depths] + ["Maximum"]

plt.figure(figsize=(8, 5))
sns.lineplot(
    data=results1,
    x="max_depth",
    y="OOB_mean",
    hue="Dataset",
    marker="o",
    linewidth=2,
)

plt.title("Out-of-bag Accuracy vs Tree Depth", fontsize=14)
plt.xlabel("Max Tree Depth", fontsize=12)
plt.ylabel("Mean OOB Accuracy", fontsize=12)
plt.legend(title="Dataset")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()

plt.xticks(ticks=range(len(depth_order)), labels=depth_order)
plt.show()
```

```python
# Replace NaN 
results2["max_features"] = results2["max_features"].replace(np.nan, "None")

feature_order = [1, 2, 4, 10, 14,"sqrt", "log2" , "All"]

plt.figure(figsize=(8, 5))
sns.lineplot(
    data=results2,
    x="max_features",
    y="OOB_mean",
    hue="Dataset",
    marker="o",
    linewidth=2,
    hue_order=sorted(results2["Dataset"].unique())
)

plt.title("Out-of-bag Accuracy vs Number of Features", fontsize=14)
plt.xlabel("Number of Features Considered at Split", fontsize=12)
plt.ylabel("Mean OOB Accuracy", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

plt.xticks(ticks=range(len(feature_order)), labels=feature_order)

plt.tight_layout()
plt.legend(title="Dataset")
plt.show()
```

```python
# Sort and select top 3 per dataset
top3 = (
    results3.sort_values(["Dataset", "OOB_mean"], ascending=[True, False])
    .groupby("Dataset")
    .head(3)
    .reset_index(drop=True)
)

top3
```
